Пары в строке

На вход поступает строка. Парой будем называть два подряд идущих символа в строке. Определите наиболее часто входящую пару в строку. В случае если таких пар несколько выведите большую из них в лексикографическом порядке.

Вормат входных данных
В единственной строке содержится исходная последовательность символов, содержащая не менее двух символов и состоящая из строчных латинских букв.

Формат выходных данных
В единственно строке необходимо вывести пару символов, которая наиболее часто встречается в данной строке.

Для примера:

Ввод	Результат
aab     ab
aabbb   bb

```python
inputStr = "vvisyuiogxelornvfupebcesivhcjnbotzumojfyefghyarxpivbvrvosboryaafqgtjlqwgsizbyeaerknrbuxxktazgcknlzbhcbdeylnyjywwqwaycjkcjrdcuhmnipgxmgmycfvreamxxgdgriqfaikoliatfrwnemeveatlpydhzkyepegvcvyuunluvzolpkksbtlopzogxjvplmiklcnorydreksmieildciillkulupoxbetqwfvzgckbkklqjynnrqicoeevcnzkmeyphvzkewhluyuoixhaoovxstmshwifxgkyrvbmhspbtcvndtlcmkxzpnqpiahskydzfnwgrpdhpnxjlec"
i = 1
pairs = {}
maxValue = 0
maxKey = ""

while i < len(inputStr):
    strKey = "{0}{1}".format(inputStr[i-1], inputStr[i])
    if strKey in pairs:
        pairs[strKey] += 1
        if maxValue < pairs[strKey]:
            maxValue = pairs[strKey]
    else:
        pairs.update({strKey: 1})
        if maxValue < 1:
            maxValue = 1
    i += 1

for k, v in pairs.items():
    # print("k: {0}, v: {1}".format(k, v))
    if v == maxValue and k > maxKey:
        maxKey = k
# print("{0}: {1}".format(maxKey, maxValue))
print("maxValue: {0}".format(maxValue))
print("maxKey: " + maxKey)
```
